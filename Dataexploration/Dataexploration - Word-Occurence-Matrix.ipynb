```python
from gensim.models import FastText
import numpy as np
path = r'G:\Fasttext\cc.en.300.bin\\cc.en.300.bin'
def get_compare_vectors(w1, w2):
    vectors = FastText.load_fasttext_format(path, encoding='utf-8')
    ev1=vectors.wv[w1]
    ev2 = vectors.wv[w2]
    print(type(ev1))
    print(ev1==ev2)
    print(np.array_equal(ev1, ev2))
```

```python
get_compare_vectors('nigger', 'niggah')
```
